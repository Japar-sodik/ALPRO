Selain PCA yang dapat digunakan untuk reduksi dimensi pada regresi maupun klasifikasi, terdapat metode lainnya yang digunakan untuk reduksi dimensi, yaitu LDA dan t-SNE.

# LDA (Linear Discriminat Analysis)

LDA adalah metode reduksi dimensi yang digunakan untuk mereduksi data sambil melakukan separasi/perpisahan antar tabel/kelas. LDA berbeda dengan PCA dimana LDA hanya dapat digunakan untuk klasifikasi, dimana membutuhkan tabel(encoding). LDA sering digunakan untuk klasifikasi dan visualisasi

Kelebihan:
*   Menggunakan Label kelas untuk memaksimalkan separasi antar kelas
*   Cocok untuk klasifikasi dan reduksi dimensi antar kelas
*   Lebaik baik PCA jika tujuannya untuk membedakan kelas



Kekurangan:
* Membutuhkan Label, tidak cocok untuk regresi/klasifikasi
* Tidak bekerja optimal untuk data non-linear

# t-SNE (t-Distrbute Stochastic Neighbor Embedding)

t-SNE merupakan metode reduksi dimensi yang **paling** cocok untuk visualisasi data 2 Dimensi dan 3 Dimensi. Metode ini tidak cocok untuk klasifikasi maupun regresi/prediksi.

Kelebihan:
*   Sangat baik untuk visualisasi data non linear
*   Ideal untuk dataset yang kompleks dan clustering
*   Memperhatikan struktur data (hubungan antar kelas)

Kekurangan:
* Hasil bisa berubah-ubah walaupun datanya sama (kecuali random state diatur)
* Waktu komputasi lebih lama dibandingkan PCA dan LDA
* Tidak cocok untuk prediksi dan regresi

# Perbandingan Visualisasi Data Iris menggunakan PCA, LDA, dan t-SNE

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data #fitur
y = iris.target #label
```

```python
y
```

```python
print(f'Data Sebelum reduksi dimensi\n {x.shape}')
```

```python
#1. Reduksi Dimensi Menggunakan PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)

#visualisasi PCA
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_pca[:,0], y = x_pca[:,1], hue=y, palette = 'Set1', legend='full')
plt.title('PCA = Reduksi dimensi dari 4D ke 2D')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()

#2. Reduksi Dimensi Menggunakan LDA
lda = LDA(n_components=2) #jadi 2 fitur
x_lda = lda.fit_transform(x, y)

#visualisasi LDA
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_lda[:,0], y = x_lda[:,1], hue=y, palette = 'Set1', legend='full')
plt.title('LDA = Reduksi dimensi dari 4D ke 2D')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')
plt.show()

#3. Reduksi Dimensi Menggunakan t-SNE
tsne = TSNE(n_components=2, random_state = 0) #jadi 2 fitur
x_tsne = tsne.fit_transform(x)

#visualisasi t-SNE
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_tsne[:,0], y = x_tsne[:,1], hue=y, palette = 'Set1', legend='full')
plt.title('t-SNE = Reduksi dimensi dari 4D ke 2D')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.show()
```
